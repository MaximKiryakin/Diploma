```python
import numpy as np
import scipy.sparse as sp
import matplotlib.pyplot as plt
from scipy.signal import convolve2d

# Создание разреженной матрицы (замените данными вашей матрицей)
data = np.ones((1000, 1000))*10
sparse_matrix = sp.csr_matrix(data)

# Создание массива с возрастами людей
ages = np.array([25, 30, 35])

# Сортировка матрицы и массива по возрасту
sorted_indices = np.argsort(ages)
sorted_matrix = sparse_matrix[sorted_indices][:, sorted_indices]

# Применение сглаживания к матрице
kernel = np.array([[1, 1, 1],
                   [1, 1, 1],
                   [1, 1, 1]]) / 9  # Фильтр для сглаживания
smoothed_matrix = convolve2d(sorted_matrix.toarray(), kernel, mode='same')

# Отображение тепловой карты с сглаживанием
plt.imshow(smoothed_matrix)
plt.xlabel('Возраст')
plt.ylabel('Возраст')
plt.title('Сглаженная тепловая карта контактов')
plt.colorbar()
plt.show()
```
